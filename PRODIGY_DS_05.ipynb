```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Example DataFrame creation (make sure df_num is properly defined)
df_num = pd.DataFrame({
    'A': [1, 2, 3, 4, 5],
    'B': [5, 4, 3, 2, 1],
    'C': [2, 3, 4, 5, 6]
})

# Check if df_num is empty
if df_num.empty:
    print("DataFrame is empty.")
else:
    # Calculate the correlation matrix
    corr_matrix = df_num.corr()

    # Check for NaN values and handle them
    if corr_matrix.isna().any().any():
        print("NaN values found in the correlation matrix. Handling NaNs...")
        corr_matrix = corr_matrix.fillna(0)

    # Plot heatmap
    plt.figure(figsize=(15, 9))
    sns.heatmap(corr_matrix, annot=True)
    plt.show()
```

```python
import matplotlib.pyplot as plt

# Assuming df1 is defined and has the necessary columns
# Create accidents_severity
try:
    accidents_severity = df1.groupby('Severity').count()['ID']
except KeyError:
    print("Ensure that 'Severity' and 'ID' columns are present in df1.")
    accidents_severity = pd.Series()  # Define as empty Series to avoid further errors

# Check if accidents_severity is defined and not empty
if not accidents_severity.empty:
    fig, ax = plt.subplots(figsize=(8, 6), subplot_kw=dict(aspect="equal"))

    # Define labels for the pie chart
    labels = accidents_severity.index.tolist()
    
    # Create pie chart
    plt.pie(accidents_severity, labels=labels,
            autopct='%1.1f%%', pctdistance=0.85)
    
    # Draw a circle in the center to make it look like a donut chart
    circle = plt.Circle((0, 0), 0.5, color='white')
    p = plt.gcf()
    p.gca().add_artist(circle)
    
    # Set title and layout
    ax.set_title("Accidents by Severity", fontdict={'fontsize': 16})
    plt.tight_layout()
    
    # Show plot
    plt.show()
else:
    print("No data available to plot.")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample data creation
df = pd.DataFrame({
    'weather': ['Clear', 'Rain', 'Fog', 'Clear', 'Rain', 'Clear', 'Snow', 'Fog', 'Clear', 'Rain', 'Snow', 'Clear']
})

# Compute weather conditions count
weather_conditions = df['weather'].value_counts()

# Plot the top 20 weather conditions
fig, ax = plt.subplots(figsize=(8, 5))
weather_conditions.sort_values(ascending=False)[:20].plot(kind='bar', ax=ax)
ax.set(title='Weather Conditions at Time of Accident Occurrence',
       xlabel='Weather',
       ylabel='Accidents Count')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Create a dummy DataFrame
test_df = pd.DataFrame({
    'Start_Lat': [34.05, 36.16, 40.71, 37.77],
    'Start_Lng': [-118.24, -115.15, -74.01, -122.42]
})

# Generate the joint plot
sns.jointplot(x='Start_Lat', y='Start_Lng', data=test_df, height=10)

# Set axis labels
plt.xlabel('Start Longitude', fontsize=12)
plt.ylabel('Start Latitude', fontsize=12)

# Show the plot
plt.show()

```
